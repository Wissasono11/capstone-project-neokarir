# 📄 Smart CV Analyzer: Tahap 4 - Konversi ke CoNLL Format

Notebook ini digunakan untuk mengonversi dataset `.jsonl` yang sudah di-tag secara otomatis ke dalam format **CoNLL-2003**. 

## 1. Import Library
Kita hanya membutuhkan library `json` bawaan Python untuk membaca file JSONL kita.

```python
import json
```

## 2. Proses Konversi JSONL ke CoNLL
Kita akan membaca isi dari `cv_dataset_auto_tagged.jsonl`, mengambil pasangan `tokens` dan `ner_tags`, lalu menuliskannya secara vertikal (menurun) ke dalam file `.conll` yang baru.

```python
input_file = "cv_dataset_auto_tagged.jsonl"
output_file = "cv_dataset_final.conll"

try:
    # Buka file input untuk dibaca dan file output untuk ditulis
    with open(input_file, "r", encoding="utf-8") as f_in, open(output_file, "w", encoding="utf-8") as f_out:
        for line in f_in:
            data = json.loads(line)
            tokens = data["tokens"]
            tags = data["ner_tags"]
            
            # Looping untuk menulis kata dan tag berdampingan, dipisah spasi
            for token, tag in zip(tokens, tags):
                f_out.write(f"{token} {tag}\n")
                
            # Tulis 1 baris kosong murni (newline) sebagai batas antar kalimat
            f_out.write("\n")
            
    print(f"✅ Konversi Sukses! Data format CoNLL berhasil disimpan di: {output_file}")

except FileNotFoundError:
    print(f"❌ Error: File '{input_file}' tidak ditemukan. Pastikan Notebook Tahap 3 sudah dijalankan dan file ada di folder yang sama.")
```

## 3. Verifikasi Hasil
Memastikan formatnya sudah persis seperti yang diminta oleh tim AI, intip 20 baris pertama dari file `.conll` yang baru saja dicetak.

```python
print("🔍 Preview Format CoNLL (20 Baris Pertama):\n")

try:
    with open(output_file, "r", encoding="utf-8") as f_preview:
        for i in range(20):
            line = f_preview.readline()
            if not line: # Berhenti jika file sudah habis sebelum 20 baris
                break
            # Gunakan argumen end="" karena line dari file sudah membawa newline (\n)
            print(line, end="") 
            
except FileNotFoundError:
     print("File belum terbuat, silakan jalankan cell konversi di atas.")
```
