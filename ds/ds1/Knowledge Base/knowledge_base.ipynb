## 1. Persiapan Daftar Link (URL)

```python
daftar_url = [
    # --- 1. Roadmap Profesi IT ---
    "https://www.revou.co/panduan-karir/front-end-developer-roadmap",
    "https://www.dicoding.com/blog/panduan-komprehensif-roadmap-menjadi-data-scientist-profesional/",
    "https://teltics.com/blog/roadmap-menjadi-backend-developer-panduan-lengkap",
    "https://www.nobelakademi.id/blog/roadmap-untuk-uiux-design-panduan-lengkap-untuk-pemula",
    "https://www.codepolitan.com/blog/artificial-intelligence-roadmap-panduan-lengkap-dan-peluang-karir-di-2026/",
    "https://itgid.org/insight/artikel-it/roadmap-cyber-security-panduan-lengkap-dan-rekomendasi-training-terbaik-ala-itgid/",
    "https://www.codepolitan.com/blog/game-design-roadmap/",
    "https://digitalskola.com/blog/home/software-engineer-roadmap",
    "https://dibimbing.id/blog/detail/network-engineer-roadmap",

    # --- 2. Tren Tools & Framework ---
    "https://www.jagoanhosting.com/blog/web-framework/",
    "https://glints.com/id/lowongan/jenis-bahasa-pemrograman/",
    "https://www.revou.co/panduan-karir/tools-data-analytics",
    "https://www.cbn.id/id/news/Solusi%20Enterprise/gen-ai-revolution-in-it-industry",
    "https://buildwithangga.com/tips/react-vue-atau-angular-panduan-memilih-framework-frontend-untuk-pemula",

    # --- 3. Tips CV ATS Friendly ---
    "https://www.dicoding.com/blog/tips-membuat-cv-programmer-yang-menarik/",
    "https://masoemuniversity.ac.id/artikel/kesalahan-cv-fresh-graduate-teknik-informatika-yang-paling-sering-bikin-langsung-ditolak-hrd/",
    "https://glints.com/id/lowongan/kesalahan-membuat-cv/",
    "https://www.dicoding.com/blog/cara-membangun-portofolio-dalam-bidang-it-untuk-fresh-graduate/",
    "https://glints.com/id/lowongan/portofolio-programmer/",
    "https://blog.tempoinstitute.com/berita/contoh-cv-ats",
    "https://id.jobstreet.com/id/career-advice/article/contoh-cv-ats-friendly-cara-membuat-template-gratis",

    # --- 4. Persiapan Interview IT ---
    "https://glints.com/id/lowongan/pertanyaan-interview-user/",
    "https://id.jobstreet.com/id/career-advice/article/interview-user-arti-contoh-pertanyaan-tips-ciri-ciri-lolos",
    "https://id.jobstreet.com/id/career-advice/article/daftar-pertanyaan-interview-programmer-tips-menjawab",
    "https://talenthero.id/insight/tips-menghadapi-coding-test",
    "https://id.jobstreet.com/id/career-advice/article/cara-negosiasi-gaji-interview-kerja-contoh"
]

print(f"Total URL yang siap diproses: {len(daftar_url)} artikel.")
```

## 2. Eksekusi Web Scraping
Pada tahap ini, looping akan dilakukan untuk mengunjungi setiap tautan. Teks utama akan diekstrak dan disimpan ke dalam *folder* lokal `Knowledge_Base_RAG` dengan ekstensi `.txt`. 

*(Terdapat jeda acak 2-4 detik per proses untuk mencegah pemblokiran dari server web asal).*

```python
# Buat folder jika belum ada
folder_output = "Knowledge_Base_RAG"
if not os.path.exists(folder_output):
    os.makedirs(folder_output)
    print(f"[*] Folder '{folder_output}' berhasil dibuat.\n")

print("Mulai proses ekstraksi artikel...\n")
berhasil = 0

for i, url in enumerate(daftar_url):
    try:
        # Download HTML mentah
        downloaded = trafilatura.fetch_url(url)
        
        if downloaded:
            # Trafilatura akan otomatis membuang tag HTML, header, footer, dan iklan
            teks_bersih = trafilatura.extract(downloaded)
            
            if teks_bersih:
                # Penamaan file
                nama_file = f"dokumen_rag_{i+1}.txt"
                path_simpan = os.path.join(folder_output, nama_file)
                
                # Simpan teks ke file
                with open(path_simpan, "w", encoding="utf-8") as f:
                    f.write(f"SUMBER URL: {url}\n")
                    f.write("-" * 50 + "\n\n")
                    f.write(teks_bersih)
                
                print(f"[{i+1}/26] ✅ BERHASIL: {nama_file}")
                berhasil += 1
            else:
                print(f"[{i+1}/26] ❌ GAGAL: Konten utama tidak ditemukan di {url}")
        else:
            print(f"[{i+1}/26] ❌ GAGAL: Akses ditolak oleh {url}")
            
        # Jeda agar tidak dianggap serangan DDoS/Spam bot
        time.sleep(random.uniform(2, 4))
        
    except Exception as e:
        print(f"[{i+1}/26] ⚠️ ERROR: {e}")

print(f"\n🎉 PROSES SELESAI! Berhasil mengamankan {berhasil} artikel ke folder '{folder_output}'.")
```
